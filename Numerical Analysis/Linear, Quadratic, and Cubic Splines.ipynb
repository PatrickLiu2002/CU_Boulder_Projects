```python
#Very short code

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, CubicSpline
```

```python
#Generate x, y
x = np.array([1, 6, 7, 9, 12, 20]) #Array for x
y = np.array([2, 8, 6, 10, 14, 41]) #Array for y
```

```python
#Plot data points
plt.plot(x, y, 'o', label = 'Data points')
```

```python
#1-d linear interpolation:
x_interp = np.linspace(np.min(x), np.max(x), 50)

y_linear = interp1d(x, y)

#Plot it
plt.plot(x, y, 'o', label = 'Data points')
plt.plot(x_interp, y_linear(x_interp), 'red', label = '1d Interpolation')
```

```python
#Quadratic Interpolation
x_interp = np.linspace(np.min(x), np.max(x), 50)
y_quadratic = interp1d(x, y, kind = "quadratic")

#Plot it
plt.plot(x, y, 'o', label = 'Data points')
plt.plot(x_interp, y_quadratic(x_interp), 'green', label = '2d Interpolation')
```

```python
#Cubic Interpolation (unbounded)
x_interp = np.linspace(np.min(x), np.max(x), 50)
y_cubic = interp1d(x, y, kind = "cubic")

#Plot it
plt.plot(x, y, 'o', label = 'Data points')
plt.plot(x_interp, y_cubic(x_interp), 'pink', label = 'Cubic Spline')
```

```python
#Cubic with boundaries

x_interp = np.linspace(np.min(x), np.max(x), 50)
y_cubicbc = CubicSpline(x, y, bc_type = "natural")

#Plot it
plt.plot(x, y, 'o', label = 'Data points')
plt.plot(x_interp, y_cubicbc(x_interp), 'black', label = 'Cubic Spline BC')
```
